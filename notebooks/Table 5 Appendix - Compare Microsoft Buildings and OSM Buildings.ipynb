```python
import pandas as pd
import sqlite3
```

```python
def load_dataframe(column):
    con = sqlite3.connect("../data/global_urban_building_completeness.gpkg")
    query = f"""
        select 
          {column} as group_name
          ,count(distinct urban_center_id) as n
          ,Cast(round(sum(osm_building_area_sqkm_2023), 0) as integer) sum_osm_building_area_sqkm_2023
          ,Cast(round(sum(microsoft_building_area_sqkm),0) as integer) sum_microsoft_building_area_sqkm
          ,round(sum(osm_building_area_sqkm_2023) / sum(microsoft_building_area_sqkm), 2) as ratio_osm_ms
        from all_parameters_urban_centers
        where {column} is not null
        group by group_name
        order by group_name 
    """
    df = pd.read_sql_query(query, con=con)
    return df
```

```python
columns = [
    "'global'",
    "region_wb",
    "shdi_2019_class",
    "ghspop_2020_class"
]

for column in columns:
    df = load_dataframe(column)
    display(df)
```
