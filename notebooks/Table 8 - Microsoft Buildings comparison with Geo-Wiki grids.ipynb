```python
import pandas as pd
import sqlite3
```

```python
def load_dataframe():
    con = sqlite3.connect("../data/global_urban_building_completeness.gpkg")
    query = f"""
        select
            region_wb
            ,count(distinct urban_center_id) as n_urban_centers
            ,count(distinct id_left) n_geowiki_locations
            ,count(*) as n_geowiki_cells
            ,sum(tp) as tp
            ,sum(fp) as fp
            ,sum(tn) as tn
            ,sum(fn) as fn
            ,case
                when sum(tp) + sum(fn) = 0 then null
                else 1.0 * sum(tp) / (sum(tp) + sum(fn))
            end as sensitivity
            ,case
                when sum(tp) + sum(fp) = 0 then null
                else 1.0 * sum(tp) / (sum(tp) + sum(fp))
            end as precision
        from geowiki_grids_final
        where
            microsoft_available is true
        group by region_wb
        order by region_wb
    """
    df = pd.read_sql_query(query, con=con)
    return df
```
